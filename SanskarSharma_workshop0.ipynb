# Create a Python program that converts between different units of measurement.

```python
def convert_meters_to_feet(meters):
    """Converts meters to feet and prints the result.

    Parameters:
        meters (float): The value in meters to convert.
    """
    one_meter = 3.28084  # Conversion factor: 1 meter = 3.28084 feet
    feet = meters * one_meter
    print(f"{meters} meters is equal to {feet} feet.")
```

```python
def convert_feet_to_meters(feet):
    """Converts feet to meters and prints the result.

    Parameters:
        feet (float): The value in feet to convert.
    """
    one_meter = 3.28084  # Conversion factor: 1 meter = 3.28084 feet
    meters = feet / one_meter
    print(f"{feet} feet is equal to {meters} meters.")
```

```python
def kg_to_pound(kg):
    """Converts kilograms to pounds and prints the result.

    Parameters:
        kg (float): The value in kilograms to convert.
    """
    one_kg = 2.20462  # Conversion factor: 1 kilogram = 2.20462 pounds
    pound = kg * one_kg
    print(f"{kg} kilograms is equal to {pound} pounds.")
```

```python
def pound_to_kg(pound):
    """Converts pounds to kilograms and prints the result.

    Parameters:
        pound (float): The value in pounds to convert.
    """
    one_kg = 2.20462  # Conversion factor: 1 kilogram = 2.20462 pounds
    kg = pound / one_kg
    print(f"{pound} pounds is equal to {kg} kilograms.")
```

```python
def litre_to_gallons(litre):
    """Converts litres to gallons and prints the result.

    Parameters:
        litre (float): The value in litres to convert.
    """
    one_litre = 0.264172  # Conversion factor: 1 litre = 0.264172 gallons
    gallons = litre * one_litre
    print(f"{litre} litres is equal to {gallons} gallons.")
```

```python
def gallons_to_litre(gallons):
    """Converts gallons to litres and prints the result.

    Parameters:
        gallons (float): The value in gallons to convert.
    """
    one_litre = 0.264172  # Conversion factor: 1 litre = 0.264172 gallons
    litre = gallons / one_litre
    print(f"{gallons} gallons is equal to {litre} litres.")
```

```python
def main():
    """Presents a menu for unit conversions and executes the selected operation."""
    print("Welcome to the Unit Converter")
    print("1. Convert meters to feet")
    print("2. Convert feet to meters")
    print("3. Convert kilograms to pounds")
    print("4. Convert pounds to kilograms")
    print("5. Convert litres to gallons")
    print("6. Convert gallons to litres")
    choice = input("Enter your choice (1-6): ")
    try:
        if choice == "1":
            meters = float(input("Enter the value in meters: "))
            convert_meters_to_feet(meters)
        elif choice == "2":
            feet = float(input("Enter the value in feet: "))
            convert_feet_to_meters(feet)
        elif choice == "3":
            kg = float(input("Enter the value in kilograms: "))
            kg_to_pound(kg)
        elif choice == "4":
            pound = float(input("Enter the value in pounds: "))
            pound_to_kg(pound)
        elif choice == "5":
            litre = float(input("Enter the value in litres: "))
            litre_to_gallons(litre)
        elif choice == "6":
            gallons = float(input("Enter the value in gallons: "))
            gallons_to_litre(gallons)
        else:
            print("Invalid choice. Please enter a number between 1 and 6.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")
```

```python
# Execute the main function to run the unit converter program.
main()
```

#Create a Python program that performs various mathematical operations on a list of numbers.

```python
def sum_of_numbers(values):
    """Calculates and prints the sum of all numbers in a list.

    Parameters:
        values (list): A list of numerical values.

    Returns:
        float: The total sum of the numbers.
    """
    sum = 0
    for value in values:
        # Corrected indexing: directly add 'value' instead of 'values[value - 1]'
        sum += value
    print(f"The total sum of given number is {sum}")
    return sum
```

```python
def average_of_numbers(values):
    """Calculates and prints the average of numbers in a list.

    Parameters:
        values (list): A list of numerical values.
    """
    sum = sum_of_numbers(values)
    average = sum/len(values)
    print(f"Average of given number is {average}")
```

```python
def maximum_of_numbers(values):
    """Finds and prints the maximum value in a list of numbers.

    Parameters:
        values (list): A list of numerical values.
    """
    maximum = values[0]
    for value in values:
        if value > maximum:
            maximum = value
    print(f"Maximum of given number is {maximum}")
```

```python
def minimum_of_numbers(values):
    """Finds and prints the minimum value in a list of numbers.

    Parameters:
        values (list): A list of numerical values.
    """
    min_val = values[0] # Renamed variable 'min' to 'min_val' to avoid shadowing built-in function
    for value in values:
        if value < min_val:
            min_val = value
    print(f"Minimum of given number is {min_val}")
```

```python
def main():
    """Provides an interactive menu for mathematical operations on a list of numbers."""
    # ask for user input
    print("Welcome to the program that performs various mathematical operations on a list of numbers.")
    try:
        list_of_numbers = [int(x) for x in input("Enter list of numbers (space separated): ").split()]
    except ValueError:
        print("Invalid input. Please enter numbers separated by spaces.")
        return

    print("We have different operations:")
    print("1. Sum of numbers")
    print("2. Average of numbers")
    print("3. Maximum of numbers")
    print("4. Minimum of numbers")
    choice = input("Enter your choice (1-4): ")

    try:
        if choice == "1":
            sum_of_numbers(list_of_numbers)
        elif choice == "2":
            average_of_numbers(list_of_numbers)
        elif choice == "3":
            maximum_of_numbers(list_of_numbers)
        elif choice == "4":
            minimum_of_numbers(list_of_numbers)
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
main() # Execute the main function to run the mathematical operations program.
```

#Write a Python function that extracts every other element from a list, starting from the first element.

```python
def extract_every_other(lst):
    """Extracts every other element from a list, starting from the first element.

    Parameters:
        lst (list): The input list.

    Returns:
        list: A new list containing every other element.
    """
    return lst[::2]
```

```python
# Example usage of extract_every_other function.
input_value = [1, 2, 3, 4, 5]
extract_every_other(input_value)
```

#Write a Python function that returns a sublist from a given list, starting from a specified index and ending at another specified index.

```python
def get_sublist(lst, start_index, end_index):
    """Returns a sublist from a given list, starting from a specified index
    and ending at another specified index (inclusive).

    Parameters:
        lst (list): The input list.
        start_index (int): The starting index of the sublist.
        end_index (int): The ending index of the sublist (inclusive).

    Returns:
        list: The extracted sublist.
    """
    return lst[start_index : end_index + 1]
```

```python
# Example usage of get_sublist function.
input = [1,2,3,4,5,6]
get_sublist(input, 1,3)
```

#Write a Python function that extracts a list of elements in reverse order starting from the second-to-last element and skipping one element in between.

```python
def reverse_skip(lst):
    """Extracts every other element starting from the first, then reverses the resulting list.

    Parameters:
        lst (list): The input list.

    Returns:
        list: A new list with every other element, reversed.
    """
    skip_val = lst[::2]
    return skip_val[::-1]
```

```python
def reverse_skip_better(lst):
    """Extracts elements in reverse order, starting from the second-to-last element
    and skipping one element in between.

    Parameters:
        lst (list): The input list.

    Returns:
        list: A new list with elements extracted in reverse and skipped.
    """
    return lst[-2::-2]
```

1. Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.

```python
def flatten(lst):
  """Flattens a list of lists into a single list.

  Parameters:
    lst (list): A nested list (e.g., [[1, 2], [3, 4]]).

  Returns:
    list: A new list containing all elements in a single dimension.
  """
  flat = [item for i in lst for item in i]
  print(flat)

nested_list = [[1, 2], [3, 4], [5]]
flatten(nested_list)
```

2. Write a Python function that extracts a specific element from a nested list given its indices.

```python
def access_nested_elements(lst, indices):
  """Extracts a specific element from a nested list given its indices.

  Parameters:
    lst (list): The nested list.
    indices (list): A list of integers representing the path to the element.

  Returns:
    any: The element at the specified nested indices.
  """
  value = lst
  for i in indices:
    # print(i)
    value = value[i]
  return value
```

```python
lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
access_nested_elements(lst, indices)
```

3. Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

```python
def sum_nested(lst):
  """Calculates the sum of all numbers in a nested list (regardless of depth).

  Parameters:
    lst (list): The nested list to sum.

  Returns:
    int or float: The total sum of all numbers.
  """
  total = 0
  for item in lst:
    if isinstance(item, list):
      total += sum_nested(item)
    else:
      total += item
  return total
```

5. Write a Python function that finds the maximum element in a nested list (regardless of depth).

```python
def find_max(lst):
    """Finds the maximum element in a nested list (regardless of depth).

    Parameters:
        lst (list): The nested list.

    Returns:
        int or float: The maximum numerical value found in the list.
    """
    max_value = 0

    for item in lst:
        if isinstance(item, list):
            value = find_max(item)
        else:
            value = item

        if value > max_value:
            max_value = value

    return max_value
```

```python
lst = [[1, 2], [3, [4, 5]], 6]
find_max(lst)
```

6. Write a Python function that counts how many times a specific element appears in a nested list.

```python
def count_occurance(lst, elem):
  """Counts how many times a specific element appears in a nested list (regardless of depth).

  Parameters:
    lst (list): The nested list.
    elem (any): The element to count.

  Returns:
    int: The total count of the element in the nested list.
  """
  count = 0
  for item in lst:
    if isinstance(item, list):
      count += count_occurance(item, elem) # Recursively count in sublists
    elif item == elem:
      count += 1
  return count
```

```python
lst = [[1, 2], [2, 3], [2, 4], 2]
elem = 2
count_occurance(lst, elem)
```

7. Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

```python
def deep_flatten(lst):
  """Flattens a list of lists of lists into a single list, regardless of the depth.

  Parameters:
    lst (list): The nested list to flatten.

  Returns:
    list: A new list containing all elements in a single dimension.
  """
  flat_list = []
  for item in lst:
    if isinstance(item, list):
      flat_list.extend(deep_flatten(item))
    else:
      flat_list.append(item)
  return flat_list
```

```python
lst = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
deep_flatten(lst)
```

8. Write a Python function that calculates the average of all elements in a nested list.

```python
def average_nested(lst):
  """Calculates the average of all elements in a nested list (regardless of depth).

  Parameters:
    lst (list): The nested list containing numerical values.

  Returns:
    float: The average of all numbers in the nested list.
  """
  flat_list = deep_flatten(lst) # Use the deep_flatten function
  if not flat_list:
      return 0.0 # Handle empty list case to avoid division by zero
  total_sum = sum(flat_list)
  average = total_sum / len(flat_list)
  return average
```

```python
lst = [[1, 2], [3, 4], [5, 6]]
average_nested(lst)
```

# Experiment: How Fast is Numpy?

```python
import time # Import the time module to measure execution time.

# Define a large size for lists and arrays to highlight performance differences.
size = 1_000_000

# Create two large Python lists.
list1 = list(range(size))
list2 = list(range(size))

# Create two equivalent NumPy arrays.
arr1 = np.array(list1)
arr2 = np.array(list2)
```

```python
# Compare the time taken for element-wise addition using Python lists and NumPy arrays.
start = time.time()
[ a + b for a, b in zip(list1, list2) ]
print("Python list addition:", time.time() - start)

start = time.time()
arr1 + arr2
print("NumPy addition:", time.time() - start)
```

```python
# Compare the time taken for element-wise multiplication using Python lists and NumPy arrays.
start = time.time()
[ a * b for a, b in zip(list1, list2) ]
print("Python list multiplication:", time.time() - start)

start = time.time()
arr1 * arr2
print("NumPy multiplication:", time.time() - start)
```

```python
# Compare the time taken for dot product calculation using Python lists and NumPy arrays.
start = time.time()
sum(a * b for a, b in zip(list1, list2))
print("Python list dot product:", time.time() - start)

start = time.time()
np.dot(arr1, arr2)
print("NumPy dot product:", time.time() - start)
```
